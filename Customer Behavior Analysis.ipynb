Create Age-Grouped Column

```python
labels = ['Young_Adult' ,'Adult','Middle_aged','Senior']
```

```python
df['age_group'] = pd.qcut(df['age'],q=4 , labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

create frequency_of_purchases columns

```python
frequency_mapping = {
    'Fortnightly' : 14 , 
    'Weekly' : 7 , 
    'Monthly' : 30 , 
    'Quarterly' :90 , 
    'Bi-Weekly' :14 ,
    'Annually' : 365 ,
'Every 3 Months' :90
}
df['purchase_frequency-days'] = df['frequency_of_purchases'].map(frequency_mapping)
    
```
